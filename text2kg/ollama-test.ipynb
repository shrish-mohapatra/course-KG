````python
from langchain_community.document_loaders.csv_loader import CSVLoader
from langchain_community.document_loaders import DirectoryLoader

# Loading transcripts folder
course = "COMP 2406"
transcript_folder = "/home/student/course-materials/COMP2406-W20/Lecture Captions"
file_ext = ".mp4.csv"
csv_loader_kwargs = {}
loader = DirectoryLoader(
    transcript_folder,
    glob="**/*.csv",
    loader_cls=CSVLoader,
    loader_kwargs=csv_loader_kwargs,
)

docs = loader.load()
print(f"Loaded {len(docs)} docs")
````

````python
# Extract transcript text
docs_to_parse = []

for doc in docs:
    doc_colummns = doc.page_content.split("\n")
    source = doc.metadata["source"]
    lecture_name = source[len(transcript_folder)+1:-len(file_ext)]
    transcript = doc_colummns[3][6:]
    print(source)
    docs_to_parse.append({"lecture_name": lecture_name, "transcript": transcript})
````

````python
import json

class KnowledgeGraph:
    def __init__(self):
        self.kg = {
            "nodes": [{"id": course}],
            "edges": [],
        }
        self.stats = {
            "num_fails": 0,
            "num_updates": 0,
        }

    def update(self, llm_output: str):
        start_delim = "{"
        start_index = llm_output.find(start_delim)
        end_delim = "}"
        end_index = llm_output.rfind(end_delim)
        json_string = llm_output[start_index:end_index + len(end_delim)]
        
        try:
            new_kg = json.loads(json_string)
        except Exception as e:
            print("BAD JSON", json_string)
            print(e)
            self.stats["num_fails"] += 1
            return
            
        print("NEW KG:", new_kg)

        for field in ["nodes", "edges"]:
            for element in new_kg[field]:
                if element not in self.kg[field]:
                    self.kg[field].append(element)
                    
        print("CURRENT KG", self.kg)
        self.stats["num_updates"] += 1


sample_llm_output = """## Knowledge Graph for Lecture Transcript "Express-P2" ```json { "nodes": [ { "id": "JSON" }, { "id": "Express-P2" }, { "id": "Built-in Middleware" }, { "id": "Pass JSON" } ], "edges": [ { "source": "JSON", "target": "Express-P2" }, { "source": "Express-P2", "target": "Built-in Middleware" }, { "source": "Built-in Middleware", "target": "Pass JSON" } ] } ``` **Nodes:** * **JSON:** The concept of JSON data format. * **Express-P2:** The lecture topic. * **Built-in Middleware:** The concept of built-in middleware in Express.js. * **Pass JSON:** The ability to pass JSON data through middleware. **Edges:** * **JSON to Express-P2:** JSON is related to Express-P2 as it is discussed in the context of the lecture topic. * **Express-P2 to Built-in Middleware:** Express-P2 is related to built-in middleware as it introduces the concept of middleware and built-in middleware specifically. * **Built-in Middleware to Pass JSON:** Built-in middleware is related to Pass JSON as it allows for passing JSON data through middleware."""
sample_bad_llm_output = """## Knowledge Graph for Lecture Transcript "Express-P2" ```json { "nodes": [ { "id": "JSON" }, { "id": "Express-P2" }, { "id": "Built-in Middleware" }, { "id": "Pass JSON" } ], "edges": [ { "source": "JSON" "target": "Express-P2" }, { "source": "Express-P2", "target": "Built-in Middleware" }, { "source": "Built-in Middleware", "target": "Pass JSON" } ] } ``` **Nodes:** * **JSON:** The concept of JSON data format. * **Express-P2:** The lecture topic. * **Built-in Middleware:** The concept of built-in middleware in Express.js. * **Pass JSON:** The ability to pass JSON data through middleware. **Edges:** * **JSON to Express-P2:** JSON is related to Express-P2 as it is discussed in the context of the lecture topic. * **Express-P2 to Built-in Middleware:** Express-P2 is related to built-in middleware as it introduces the concept of middleware and built-in middleware specifically. * **Built-in Middleware to Pass JSON:** Built-in middleware is related to Pass JSON as it allows for passing JSON data through middleware."""
kg = KnowledgeGraph()
# kg.update(sample_bad_llm_output)

kg.update(sample_llm_output)
kg.update(sample_llm_output)
````

````python
from langchain import hub
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.pydantic_v1 import BaseModel, Field

class Node(BaseModel):
    id: str = Field(description="key concept introduced in a lecture")

class Edge(BaseModel):
    source: str = Field(description="source Node id")
    target: str = Field(description="target Node id")

class KnowledgeGraph(BaseModel):
    nodes: List[Node] = Field(description="list of nodes representing concepts")
    edges: List[Edge] = Field(description="list of edges representing prequisite relationship between concepts")

system_prompt = "You are an assitant which is creating a knowledge graph of key concepts from lecture transcripts of a computer science course."
current_kg_prompt = "Currently the knowledge graph contains the following nodes & edges:"
# parser = JsonOutputParser(pydantic_object=KnowledgeGraph)

format_instructions = "{nodes: [{id: <name of concept>}...], edges: [{source: <id of source concept node>, target: <id of target concept node>}]}"

prompt = PromptTemplate(
    template="{system_prompt}\n{current_kg_prompt}\n{current_kg}\n{format_instructions}\n{transcript}\n",
    input_variables=["current_kg", "transcript"],
    partial_variables={
        "system_prompt": system_prompt,
        "current_kg_prompt": current_kg_prompt,
        "format_instructions": format_instructions,
        # "format_instructions": parser.get_format_instructions()
    },
)

# prompt = hub.pull("hwchase17/react")
print(prompt)
````
